## Mutual Information

Here we first calculate the mutual information of 2 Bournuli Random Variable. And then calculate the mutual info of two inhomeogeneous spike train of firing rate determined by LN model.

$$\Phi=m_A(t_A)m_B(t_B)\sum^\infty_{n=2}\frac{(-\mu)^n}{n(n-1)}dt_Adt_B+\mathbb{O}(dt^3)$$
In the equation, $\mu$ is the firing rate's response coefficient to perturb in upstream firing events. $m_B^\delta(t_B)$ is the perturbed firing rate in B due to a firing event in A at $t_A$.
$$\mu(t_A,t_B)=\frac{m_B^\delta(t_B)-m_B(t_B)}{m_B(t_B)}$$

The equation is obtained in weak coupling regime, thus perturbation is viable. 

### Analysis
* Firstly, perturbation coefficient is restricted by casuality and "speed" of interaction, thus $\mu(t_A,t_B)==0$ for $t_A\geq t_B$
* Aside from the quadratic dependence on perturb coef $\mu$, $\Phi$, linearly depend on $m_A(t_A),m_B(t_B)$ at the 2 time points, i.e. the higher rate the 2 population fires, the higher mutual information it is. 
* Mutual information is symmetric by definition, so **how to interpret this Eq in symmetrical way**?

For a synfire chain, we consider the mutual information of 2 successive layers/populations. A and B, which has a voltage distribution function $\rho_i(V,t)$, 1st moment $M^1_i(t)$, firing rate $m_i(t)$, current drive $I_i(t)$ with $i=A,B$. 

## Voltage Field Equation (Fokker Planck Description)

The gaussian model is developed as an approximation to the Fokker Planck description of homeogeneous neuron popultion. 
$$\partial_t \rho(V,t)=- \partial_V J(V,t)=\partial_V[(g_L(V-V_R)-I(t))\rho(V,t)]+\frac{\sigma^2(t)}{2}\partial_V^2\rho(V,t)$$
The current defined as
$$J(V,t)=[-g_L(V-V_R)+I(t)-\frac{\sigma^2(t)}{2}\partial_V]\rho(V,t)$$
Especially, the diffusion term $\frac{\sigma^2(t)}{2}\partial_V\rho(V,t)$, has the effective diffusivity is consist of variance of background input (or gating current) $\sigma_0^2$, and upstream spikes rate $m_A$
$$\frac{\sigma^2(t)}{2}=\sigma_0^2+\frac12 \frac{S^2}{N}m_A(t)$$

Ouput firing rate and the input current has the relation,
$$I_B +\tau\frac{d}{dt}I_B = S m_A(t)$$
In another word, the $I_B(t)$ is the firing rate convolving with a time constant $\tau$ (of exponential decay). 
$$I_B(t)=\exp(-t/\tau)\theta(t) * S m_A(t)=\int^t_0 S m_A(s)\exp(-(t-s)/\tau)$$

This time constant $\tau$ signify the decay time constant of synaptic channel. 

Note that increase of upstream firing rate has mainly **2 effects**, namely 
* increase of current drive $I(t)$ 
* increase of diffusivity $\sigma^2(t)$.

And the 2nd effect can be ignored only if the coupling $S$ is weak enough, or enough background current noise overwhelms the change. 

### Perterbation propagation
So in this simplification, the causal chain of perturbation is: 
* Spike events in A pop at $t_A$ caused a increase in B pop current input $I_B$. And the response function is EPSC, look like alpha function. Denote with $\kappa(t-t_A)$
* The current perturbation's effect accumulate, as it interfere with Voltage probability distribution field's evolution. And the EPSC's influence on the field evolution is the hardest part of problem.
* The final state of evolution $\rho(V,t_B)$ and the drive $I_B(t_B)$ together determine the marginal current $J(V_{Th},t_B)$, i.e. firing rate. 


### Gaussian simplification
In the Gaussian simplification, we restricted the distribution in the gaussian family (with cutoff). 
$$\rho(V,t)=\frac1P \exp(-\frac{(V-\mu(t))^2}{\sigma(t)^2})$$
The gaussian shape is obtained by solving the steady state of Voltage field equation (with constant drive) in a much larger field.
$$J(V)=[-g_L(V-V_R)+I-\frac{\sigma^2}{2}\partial_V]\rho(V)=Const$$

If we further ignore the change of bump width, then the only possible motion is translation. The distribution function is
$$\rho(V,t)=\frac1P \exp(-\frac{(V-M_B^1(t))^2}{\sigma^2})$$
And $\sigma^2=\sigma^2(t)/2g_L$, $\sigma^2(t)/2=\sigma_0^2+\frac{S^2m_A(t)}{2N}$. And we use the first term to approximate

Thus we have moment equation, 
$$\frac{d}{dt}M_B^1(t)=-g_LM_B^1(t)+I(t)$$
The solution to the Eq. is 
$$M_B^1(t)=v_0\exp(-g_Lt)+ \exp(-g_Lt)\theta(t)*I(t)\\=v_0\exp(-g_Lt)+\int^t_0 \exp(-g_L(t-s))I(s)ds$$
The 1st term is the solution of the homeogeneous linear ODE, and the 2nd term is the drive convolve with the time constant (Green function). This time constant $g_L$ signifies the membrane time constant that filter the input current.

In this view, a spike in upstream population experience the filter/convolution of 2 time constants, channel $\tau$ and membrane $g_L$. And it has the result of alpha function shape. 

Firing rate or marginal flow's dependency on $M^1$ and $I$ goes like, 
$$m_B(t)=\frac1P \exp(-\frac{2g_L(V-M^1_B(t))^2}{\sigma^2(t)}) \cdot (I(t)-g_L M_B^1(t))\\
\approx \frac1P \exp(-\frac{g_L(V-M^1_B(t))^2}{\sigma_0^2}) \cdot (I(t)-g_L M_B^1(t))$$

**Note** this expression has the assumption that the $M1$ has not been pushed too hard to cross $V_Th$. 
If it's not the case, we must consider the recurring density at $V_r$

Assuming gaussian shape of Voltage configuration, the firing rate and 1st Moment has a simple dependency.
$$$$
And firing rate is just the marginal (at $V_{Th}$) current $m(t)=J(V_{Th},t)$
$$$$
And 1st Moment has simple relation to $I_A(t)$ as well.


## Graded Transfer in Gaussian model

Before, we focus on graded transfer with Synfire Chain, i.e. order preversing property of the "map" $I_A(t)\rightarrow I_B(t)$. This property corresponds to an line attractor (in the phase space of the iterative map system) as its underpinning. 



```python
import numpy as np
import matplotlib.pyplot as plt
#from scipy.optimize import curve_fit
#from numpy import inf
```

```python
def Moment1(times,A=1,Ig=1,v0=0.5,tau=5,gL=0.1):
    '''Assume times begin at 0'''
    M1=v0*np.exp(-gL*times) + Ig/gL*(1-np.exp(-gL*times))+A/(1/tau-gL)*(np.exp(-gL*times)-np.exp(-times/tau))
    return M1
```

```python
times=np.arange(0,100,0.1)
plt.figure(figsize=[10,4])
plt.plot(times,Moment1(times,A=1,Ig=0.01,v0=0.5,tau=5,gL=0.1))
plt.title("Dynamics of Voltage distribution mean")
plt.show()
```

```python
from scipy.special import erf
def NormDist(V,M1,v0=0.5,vTh=1,sigma2=0.1,gL=0.1):
    NormCoef=erf((vTh-v0)/np.sqrt(sigma2/gL))/2 + 0.5 # Integral of normal distribution with cutoff
    #print(NormCoef)
    rho=1/NormCoef / np.sqrt(np.pi*sigma2/gL)*np.exp(-gL*(V-M1)**2/sigma2)
    return rho
```

```python
plot=True
```

```python
vTh=1;gL=0.05;tau=5
A=0.1;Ig=0.05;v0=0.1;sigma2=0.01
S=0.1
dt=0.01;
times=np.arange(0,50,dt)

M1=Moment1(times,A=A,Ig=Ig,v0=v0,tau=tau,gL=gL)
if M1.max()>1*vTh:
    print("Warning!! Gaussian Approximation invalid")
rho_marg=NormDist(vTh,M1,v0=v0,vTh=vTh,sigma2=sigma2,gL=gL)
IB=A*np.exp(-times/tau)
Jmarg=(IB+Ig-gL*M1)*((IB+Ig-gL*M1)>0)*rho_marg
WindF=np.exp(-times/tau)
IOut=np.convolve(Jmarg,WindF)*dt*S
IOut=IOut[:times.size]
if plot:
    plt.figure(figsize=[10,13])
    plt.subplot(511)
    plt.plot(times,M1)
    plt.ylabel(r"1st Moment $M^1_B$")
    plt.subplot(512)
    plt.plot(times,rho_marg)
    plt.ylabel(r"Marginal density $\rho(V_{Th})$")
    plt.subplot(513)
    plt.plot(times,IB+Ig-gL*M1)
    plt.ylabel("Drive: IB+Ig-gL*M1")
    plt.subplot(514)
    plt.plot(times,Jmarg)
    plt.ylabel(r"Marginal Current $m_B$")
    plt.subplot(515)
    plt.plot(times,IOut)
    plt.ylabel(r"Output Current $I_out$")
    plt.show()
```

## MI Cauculation

By Zhuocheng's Calculation, the perturb coefficient goes like, 
$$\mu(t_A,t_B)=\frac SN\frac{1}{gL-1/\tau}\{\frac{2gL}{\sigma_0^2}(V_{th}-M_B^1(t_B))(\exp(-gL Dt)-\exp(-Dt/\tau)) + \frac{gL\exp(-gL Dt)-\exp(-Dt/\tau)/\tau}{I(t_B)-gL M_B^1(t_B)}\}$$
$Dt=t_B-t_A$. 

### Analysis
Recall the firing rate equation as product of 2 terms : 
$$\frac1P \exp(-\frac{g_L(V-M^1_B(t))^2}{\sigma_0^2}) \cdot (I(t)-g_L M_B^1(t))==\rho(V_{Th},t_B)\frac{d}{dt}M_B^1(t)$$
The 2 terms in $\mu$ corresponds to the 2 ways that a current perturb can affect the output firing rate (marginal flow). 

* Change the density at edge, through changing the position of distribution mean $M^1(t_B)$. 
    * The closer $M_B^1(t_B)$ is to $V_{th}$, the flatter the distribution is. Then the density at edge is less sensitive to change of mean $M_B^1$. 
    * And the EPSP/alpha function like term $\exp(-gL Dt)-\exp(-Dt/\tau)$ describe the Voltage response as a function of time.
* Change the speed term $I(t_B)-g_L M_B^1(t)$, through increase of current and move of 
    * Inverse proportional to the current move speed $I(t_B)-g_L M_B^1(t)$, the smaller it is, the more sensitive it is to a small perturb $\propto S/N$
    * Due to this when $I(t_B)-g_L M_B^1(t)\approx 0$ the second term may provide some **singularity**!! (*Note that at the singular point perturbation theory will FAIL!!*)
    
And we find the overall coefficient: 
* $S/N$ describe the single synaptic strength, or perturb strength. Smaller the perturb, the mutual information cutoff to 1st order will be smaller. 
* $\frac{1}{gL-1/\tau}$ is a coefficient to normalize amplitude of alpha functions. 

```python
def PertubCoef(tA,tB,M1,I,S=0.1,N=100,tau=5,gL=0.1,sigma2=0.01,vTh=1):
    Dt=tB-tA
    mu=S/N/(gL-1/tau)*( 2*gL/sigma2*(vTh-M1)*(np.exp(-Dt/tau)-np.exp(-gL*Dt)) \
                       +(gL*np.exp(-gL*Dt)-np.exp(-Dt/tau)/tau) / (I-gL*M1))
    return mu*(tB>tA)
```

```python
vTh=1;gL=0.05;tau=5
A=0.1;Ig=0.05;v0=0.1;sigma2=0.01
S=0.1
dt=0.01;
times=np.arange(0,50,dt)

#idx=1000
#tB=times[idx]
tA=0
M1=Moment1(times,A=A,Ig=Ig,v0=v0,tau=tau,gL=gL)
if M1.max()>1*vTh:
    print("Warning!! Gaussian Approximation invalid")
rho_marg=NormDist(vTh,M1,v0=v0,vTh=vTh,sigma2=sigma2,gL=gL)
IB=A*np.exp(-times/tau)
Jmarg=(IB+Ig-gL*M1)*((IB+Ig-gL*M1)>0)*rho_marg
WindF=np.exp(-times/tau)
IOut=np.convolve(Jmarg,WindF)*dt*S
IOut=IOut[:times.size]

#PertubCoef(tA,tB,M1[idx],(IB[idx]+Ig-gL*M1[idx]))
mu=PertubCoef(tA,times,M1,(IB+Ig-gL*M1),S=S,tau=tau,vTh=vTh,sigma2=sigma2,gL=gL)
MI=A*Jmarg*mu**2/2
```

```python
plt.figure(figsize=[10,8])
plt.subplot(211)
plt.plot(times,mu)
plt.ylim(-0.0,0.05)
plt.ylabel(r"Perturb Coef $\mu$")
plt.subplot(212)
plt.plot(times,MI)
plt.ylim(0,10E-7)
plt.ylabel("Instantaneous MI")
plt.show()
```
